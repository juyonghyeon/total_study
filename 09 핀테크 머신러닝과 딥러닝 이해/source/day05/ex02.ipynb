```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1000, 1000, 0.1)
y = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = 1 / (1 + np.exp(-x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
